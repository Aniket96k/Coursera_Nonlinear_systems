```python
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
data_A = np.random.rand(10)
data_B = np.random.rand(10)
data_C = np.random.rand(10)

# Display data
data_A, data_B, data_C
```

```python
# Calculate summary statistics
mean_A = np.mean(data_A)
mean_B = np.mean(data_B)
mean_C = np.mean(data_C)

std_A = np.std(data_A)
std_B = np.std(data_B)
std_C = np.std(data_C)

# Display summary statistics
mean_A, std_A, mean_B, std_B, mean_C, std_C
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data_A, label='A', marker='o')
plt.plot(data_B, label='B', marker='x')
plt.plot(data_C, label='C', marker='s')
plt.legend()
plt.title('Sample Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
